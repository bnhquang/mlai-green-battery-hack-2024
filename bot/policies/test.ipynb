```python
import pandas as pd
from collections import deque
```

```python
df = pd.read_csv('../data/training_data.csv')

df.head()

prices = df['price']

prices
```

```python
price_series = pd.Series(prices)
```

```python
price_series
```

```python
# Calculate a simple volatility index (standard deviation of price changes)
price_series = pd.Series(prices)
price_changes = price_series.diff().dropna()
volatility_index = price_changes.std()
volatility_index
```

```python
vol_list = []

for i in range(len(prices) - 283000):
    price_series = pd.Series(prices[i:i+283])
    price_changes = price_series.diff().dropna()
    volatility_index = price_changes.std()
    vol_list.append(volatility_index)
    
print(vol_list)
```

```python
import numpy as np

vol_list = np.array(vol_list)
vol_list = vol_list[vol_list <= 200]
```

```python
import matplotlib.pyplot as plt

# Create the histogram
plt.hist(vol_list, bins=30, edgecolor='black')

# Add labels and title
plt.xlabel('Volume')
plt.ylabel('Frequency')
plt.title('Histogram of Volume')

# Display the histogram
plt.show()
```
